District-wise Annual Data on Area, Yield, and Production**

This dataset is organized into two main files:

* **Area, Production, and Yield:**
      Contains annual data for 20 major crops, including cereals, pulses, oilseeds, cotton, sugarcane, and aggregated totals for fruits and vegetables.       Yield figures are derived by dividing total production by the cultivated area.

* **High Yielding Varieties:**
Provides information on the adoption and distribution of high-yielding crop varieties.

The dataset captures the yearly area and production statistics for each crop. The percentage share of each crop’s area is calculated relative to the Gross Cropped Area (GCA), which is computed using a standardized methodology.
